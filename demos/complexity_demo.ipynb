```python
from dynadojo.systems.gilpin_flows import GilpinFlowsSystem
from dynadojo.wrappers import SystemChecker
from dynadojo.utils.lds import plot
from dynadojo.utils.complexity_meas import gp_dim, mse_mv, pca, find_lyapunov_exponents, find_max_lyapunov, kaplan_yorke_dimension, pesin
```

```python
latent_dim = 3
embed_dim = latent_dim
test_size = 3
timesteps = 1000
system = SystemChecker(GilpinFlowsSystem(latent_dim, embed_dim, 'Lorenz', seed=42))
x0 = system.make_init_conds(1)
x = system.make_data(x0, timesteps=timesteps)
plot([x], labels=["X"])
```

```python
print ("corr_dim:", gp_dim(x[0]))
print("mse:", mse_mv(x[0]))
print("pca:", pca(x[0]))
model = system._system.system
lyapunov_spectrum = find_lyapunov_exponents(model, timesteps)
print("lyapunov:", lyapunov_spectrum)
print("max_lyaponuv:", find_max_lyapunov(lyapunov_spectrum))
print("ky_dim:", kaplan_yorke_dimension(lyapunov_spectrum))
print("pesin:", pesin(lyapunov_spectrum))
```
